```python
x = DS.iloc[:,0:11].values
y = DS.iloc[:,12].values
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
DS["DEATH_EVENT"]=le.fit_transform(DS["DEATH_EVENT"])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(11,kernel_initializer = "uniform", activation ="relu"))
```

```python
model.add(Dense(22,kernel_initializer = "uniform", activation ="relu"))
```

```python
model.add(Dense(44,kernel_initializer = "uniform", activation ="relu"))
```

```python
model.add(Dense(1,kernel_initializer = "uniform", activation ="sigmoid"))
```

```python
model.compile("adam","binary_crossentropy",metrics = ["accuracy"])
```

```python
model.fit(x_train,y_train,batch_size = 16, epochs=400)
```
