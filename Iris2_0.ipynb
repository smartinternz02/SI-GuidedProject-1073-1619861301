```python
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
sb.set()
```

```python
iris = sb.load_dataset("iris")
```

## SEPAL LEN VS WIDTH FOR SETOSA

```python
sb.scatterplot(x="sepal_width",y="sepal_length",data=iris.loc[iris["species"]=="setosa"])
```

## SEPAL LEN VS WIDTH VIRGINICA

```python
sb.scatterplot(x="sepal_width",y="sepal_length",data=iris.loc[iris["species"]=="virginica"])
```

## SEPAL LEN VS WIDTH VERSICOLOR

```python
sb.scatterplot(x="sepal_width",y="sepal_length",data=iris.loc[iris["species"]=="versicolor"])
iris["species"].unique()
```

## PETAL LEN VS WIDTH SINCE PETAL IS MORE VALUABLE TO FLOWER
## SEPAL LEN VS WIDTH DISTRIBUTIONS TO FIND OUT THE MOST UNIFORM FLOWER WRT TO PETAL SIZE AND SEPAL SIZE

```python
plt.figure(figsize=(16,16))
plt.subplot(2,2,1)
sb.distplot(iris["petal_length"].loc[iris["species"]=="setosa"],color="g")
plt.title("SETOSA PETAL LENGTH")
plt.subplot(2,2,2)
sb.distplot(iris["petal_width"].loc[iris["species"]=="setosa"],color="r")
plt.title("SETOSA PETAL WIDTH")
plt.subplot(2,2,3)
sb.distplot(iris["sepal_length"].loc[iris["species"]=="setosa"],color="b")
plt.title("SETOSA SEPAL LENGTH")
plt.subplot(2,2,4)
sb.distplot(iris["sepal_width"].loc[iris["species"]=="setosa"],color="y")
plt.title("SETOSA SEPAL WIDTH")
```

```python
plt.figure(figsize=(16,16))
plt.subplot(2,2,1)
sb.distplot(iris["petal_length"].loc[iris["species"]=="virginica"], color="purple")
plt.title("Virginica Petal Length")
plt.subplot(2,2,2)
sb.distplot(iris["petal_width"].loc[iris["species"]=="virginica"], color="brown")
plt.title("Virginica Petal Width")
plt.subplot(2,2,3)
sb.distplot(iris["sepal_length"].loc[iris["species"]=="virginica"],color="orange")
plt.title("Virginica Sepal Length")
plt.subplot(2,2,4)
sb.distplot(iris["sepal_width"].loc[iris["species"]=="virginica"],color="blue")
plt.title("Virginica Sepal Width")
```

```python
plt.figure(figsize=(16,16))
plt.subplot(2,2,1)
sb.distplot(iris["sepal_length"].loc[iris["species"]=="versicolor"],color="orange")
plt.title("Versicolor Sepal Length")
plt.subplot(2,2,2)
sb.distplot(iris["sepal_width"].loc[iris["species"]=="versicolor"],color="blue")
plt.title("Vericolor Sepal Width")
plt.subplot(2,2,3)
sb.distplot(iris["petal_length"].loc[iris["species"]=="versicolor"],color="green")
plt.title("Versicolor Petal Length")
plt.subplot(2,2,4)
sb.distplot(iris["petal_width"].loc[iris["species"]=="versicolor"],color="brown")
plt.title("Vericolor Petal Width")
```

## PETAL LENGTH AND WIDTH COMPARISON ACROSS DIFF SPECIES

```python
sb.catplot(x="species",y="petal_length",data=iris)
sb.catplot(x="species",y="petal_width",data=iris)
```

## SEPAL LENGTH AND WIDTH COMPARISON ACROSS DIFF SPECIES

```python
sb.catplot(x="species",y="sepal_width",data=iris)
sb.catplot(x="species",y="sepal_length",data=iris)
```

## BOXPLOT VERSIONS OF THE SAME 

```python
plt.figure(figsize=(16,16))
plt.subplot(2,2,1)
sb.boxplot(x="species",y="sepal_width",data=iris)
plt.title("SEPAL WIDTH")
plt.subplot(2,2,2)
sb.boxplot(x="species",y="sepal_length",data=iris)
plt.title("SEPAL LENGTH")
plt.subplot(2,2,3)
sb.boxplot(x="species",y="petal_length",data=iris)
plt.title("PETAL LENGTH")
plt.subplot(2,2,4)
sb.boxplot(x="species",y="petal_width",data=iris)
plt.title("PETAL WIDTH")
```

```python
plt.figure(figsize=(16,16))
plt.subplot(2,1,1)
sb.barplot(x="sepal_length",y="sepal_width",data=iris)
plt.title("SEPAL LENGTH VS WIDTH")
plt.subplot(2,1,2)
sb.barplot(x="petal_length",y="petal_width",data=iris)
plt.title("PETAL LENGTH VS WIDTH")
```
